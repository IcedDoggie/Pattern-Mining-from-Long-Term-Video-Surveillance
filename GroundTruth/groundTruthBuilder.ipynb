```python
import numpy as np
import pandas as pd

col_names = ['Date', 'Anomaly']
chingGT = pd.read_table("groundTruth ching.txt", header=None, names=col_names)
jhGT = pd.read_table("groundTruth jh.txt", header=None, names=col_names, sep=' ')
ibGT = pd.read_table("groundTruth ib.txt", header=None, names=col_names, sep=' ')

anomalyCol_ching = chingGT[['Anomaly']]
anomalyCol_jh = jhGT[['Anomaly']]
anomalyCol_ib = ibGT[['Anomaly']]

anomaly_totalup = anomalyCol_ching.add(anomalyCol_jh, anomalyCol_ib)
anomaly_totalup = anomaly_totalup.divide(2)
anomaly_totalup = anomaly_totalup.astype(int)

ibGT = ibGT.drop('Anomaly', axis=1)
ibGT['Anomaly'] = anomaly_totalup 
anomaly_totalup = ibGT


pd.set_option('display.max_rows', len(anomaly_totalup))
# print(type(anomalyCol_ching[['Anomaly']].loc[0].Anomaly))
# anomaly_totalup = pd.DataFrame.as_matrix(anomaly_totalup)
print(anomaly_totalup)

# filename = 'groundtruth_3.txt'
np.savetxt(r'E:\Documents\MMU Studies\Python Scripts\GroundTruth\groundTruth_3.txt',
           anomaly_totalup.values, fmt='%s')
```
